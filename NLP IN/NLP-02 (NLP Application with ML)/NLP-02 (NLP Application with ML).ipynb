### Cleaning URL, Tags, Mentions for Tweets

```python
import re
s = "http\\:www.mynet.com #lateflight @airlines"
s = re.sub("http\S+", "", s).strip()
s

# \S boluk hari dier tm karakterleri temsil eder.
# http\S+, http ile balayan ve http'den sonra boluk hari en az bir veya daha fazla karakter ihtiva eden paternleri temsil 
# eder.
# http\S*, http ile balayan ve http'den sonra boluk hari en az sfr veya daha fazla karakter ihtiva eden paternleri temsil
# eder.
# regex'in sub fonksiyonu kendisine verilen text iirisinde aradmz paternlerini bulup silmemize veya istediimiz bir 
# karakter/karakterlerle deitirmemizi salar.
# re.sub("http\S+", "", s) kodu s deikenine atadmz text ierisinde http ile balayan ve http'den sonra boluk hari en az
# bir veya daha fazla karakter ihtiva eden paternleri bulur ve siler.

# re.sub("http\S+", "a", s) kodu s deikenine atadmz text ierisinde http ile balayan ve http'den sonra boluk hari en az
# bir veya daha fazla karakter ihtiva eden paternleri bulur ve a karakteri ile deitirir.
```

```python
s = re.sub("#\S+", "", s)
s

# tag'i temizliyoruz. Ancak grdnz gibi text'in banda bir boluk karakteri olutu.
```

```python
s = re.sub("@\S+", "", s)
s.strip()

# mention temizliyoruz.

# \S boluk hari dier tm karakterleri temsil ettiinden yukrda grdnz zere boluklara bir ilem yapmaz. Bu sebeple 
# textimizin banda veya sonunda kalan boluk karakterlerini strip() fonksiyonu ile temizliyoruz.
```

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
        stop_words.remove(i)
```

```python
def cleaning(data):
    
    import re
    
    #1. Removing URLS
    data = re.sub('http\S*', '', data).strip()
    data = re.sub('www\S*', '', data).strip()

    #2. Removing Tags
    data = re.sub('#\S*', '', data).strip()

    #3. Removing Mentions
    data = re.sub('@\S*', '', data).strip()
    
    #4. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")
    
    #5. Tokenize
    text_tokens = word_tokenize(data.lower())
    
    #6. Remove Puncs and numbers
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #7. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #8. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
       
    #9. joining
    return " ".join( text_cleaned)
```

```python
cleaning_text = df2["text"].apply(cleaning)
cleaning_text.head()

# textimi temizleyip cleaning_text deikenine atyorum. Notebookun sonunda WordCloud grselletirmesinde kullanacaz.
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    
```
